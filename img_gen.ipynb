```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

def save_image_from_file(file, index, image_type):
    # Load the .mat file
    mat_data = scipy.io.loadmat(file)

    shape_dict = {'dots': 100,
                  'rectangles': 30}
    
    # Assuming the data is stored under 'D' in the .mat file
    data = mat_data['D']  # Adjust the key if necessary

    # Ensure the index is within the bounds of the dataset
    tot_patterns = data.shape[1]
    if index < 0 or index >= tot_patterns:
        raise ValueError(f"Index out of bounds. Please select an index between 0 and {tot_patterns - 1}.")
    
    # Display the selected image, reshaping the data (assuming 30x30 images)
    plt.imshow(data[:, index].reshape(shape_dict[image_type], shape_dict[image_type]), cmap='gray')
    
    # Adjust the axis properties
    plt.axis('off')
    plt.gca().set_aspect('equal')

    # Show the image
    plt.tight_layout()

    plt.savefig(f'/Users/oparkinson/Desktop/testin_{image_type}_code.png',pad_inches=0 ,bbox_inches = 'tight')

path_rectangles = '/Users/oparkinson/Downloads/DevSci_code 2/data/SZ_data.mat'
path_dots = '/Volumes/Expansion/macbook_files/NumStim_7to28_100x100_TrainSet.mat'
index_rectangles = 51199
index_dots = 21969
image_type_dots = 'dots'
image_type_rects = 'rectangles'
save_image_from_file(path_rectangles,index_rectangles,image_type_rects)
save_image_from_file(path_dots,index_dots,image_type_dots)
```
